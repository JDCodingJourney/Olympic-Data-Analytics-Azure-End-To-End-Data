```python
# Reload entriesgender with correct delimiter
entriesgender = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ";") \
    .load(f"{raw_data_path}/Entriesgender.csv")

print("✅ entriesgender reloaded with semicolon delimiter")
print("\nSchema after reload:")
entriesgender.printSchema()
print("\nSample data:")
entriesgender.show(10)
```

```python
# Reload all files with semicolon delimiter if needed
athletes = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ";") \
    .load(f"{raw_data_path}/Atheltes.csv")

coaches = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ";") \
    .load(f"{raw_data_path}/Coaches.csv")

entriesgender = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ";") \
    .load(f"{raw_data_path}/Entriesgender.csv")

medals = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ";") \
    .load(f"{raw_data_path}/Medals.csv")

teams = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ";") \
    .load(f"{raw_data_path}/Teams.csv")

print("✅ All data reloaded with semicolon delimiter")
print(f"\nathletes: {athletes.count()} records")
print(f"coaches: {coaches.count()} records")
print(f"entriesgender: {entriesgender.count()} records")
print(f"medals: {medals.count()} records")
print(f"teams: {teams.count()} records")
```

```python
# Verify the schema after reload
print("=== ATHLETES SCHEMA ===")
athletes.printSchema()
athletes.show(5)

print("\n=== ENTRIES GENDER SCHEMA ===")
entriesgender.printSchema()
entriesgender.show(10)

print("\n=== MEDALS SCHEMA ===")
medals.printSchema()
medals.show(5)
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType

# Convert columns to integer type
entriesgender = entriesgender.withColumn("Female", col("Female").cast(IntegerType())) \
    .withColumn("Male", col("Male").cast(IntegerType())) \
    .withColumn("Total", col("Total").cast(IntegerType()))

print("✅ Updated entriesgender schema:")
entriesgender.printSchema()
entriesgender.show(10)
```

```python
# Check the column names in medals dataframe
print("Medals columns:")
print(medals.columns)

# Display first few rows to see the structure
print("\nSample data:")
medals.show(5)
```

```python
# Calculate the average number of participants by gender for each discipline
average_entries_by_gender = entriesgender.withColumn(
    'Avg_Female', entriesgender['Female'] / entriesgender['Total']
).withColumn(
    'Avg_Male', entriesgender['Male'] / entriesgender['Total']
)

print("📊 Average participants by gender per discipline:")
average_entries_by_gender.show(20)
```

```python
# Save all transformed data to the transformed-data folder
athletes.repartition(1).write.mode("overwrite").option("header", 'true').csv(f"{transformed_data_path}/athletes")
coaches.repartition(1).write.mode("overwrite").option("header", "true").csv(f"{transformed_data_path}/coaches")
entriesgender.repartition(1).write.mode("overwrite").option("header", "true").csv(f"{transformed_data_path}/entriesgender")
medals.repartition(1).write.mode("overwrite").option("header", "true").csv(f"{transformed_data_path}/medals")
teams.repartition(1).write.mode("overwrite").option("header", "true").csv(f"{transformed_data_path}/teams")

print("✅ All data has been successfully saved to:")
print(f"📁 {transformed_data_path}")
```
